```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def series_ar1(T, φ):
    y = np.zeros(T)
    for t in range(1, T):
        y[t] = φ * y[t-1] + np.random.randn()
    return y
```

```python
T = 1_000
np.random.seed(654321)

phis = list(np.round(np.arange(0.0, 1.0, 0.1), 1)) + [0.99, 1.0]
fig, axes = plt.subplots(3, 4, figsize=(14, 9), sharex=True)
axes = axes.ravel()

for ax, φ in zip(axes, phis):
    y = series_ar1(T, φ)
    ax.plot(y, lw=1)
    ax.set_title(f"φ = {φ}", fontsize=10)
    ax.set_xlabel("t", fontsize=8)
    ax.tick_params(axis='both', labelsize=7)

fig.suptitle(f"AR(1) trajectories as φ approaches 1 for T={T:,} timesteps", fontsize=14)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
```

```python
def series_ar1(T, φ):
    y = np.zeros(T)
    for t in range(1, T):
        y[t] = φ * y[t-1] + (1 if np.random.random() < 0.5 else -1)
    return y
```

```python
T = 1_000_000
np.random.seed(654321)

phis = list(np.round(np.arange(0.0, 1.0, 0.1), 1)) + [0.99, 1.0]
fig, axes = plt.subplots(3, 4, figsize=(14, 9), sharex=True)
axes = axes.ravel()

for ax, φ in zip(axes, phis):
    y = series_ar1(T, φ)
    ax.plot(y, lw=1)
    ax.set_title(f"φ = {φ}", fontsize=10)
    ax.set_xlabel("t", fontsize=8)
    ax.tick_params(axis='both', labelsize=7)

fig.suptitle(f"AR(1) trajectories as φ approaches 1 for T={T:,} timesteps", fontsize=14)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
```
